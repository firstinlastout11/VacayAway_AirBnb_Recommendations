```python
# Definethe API key, Define the Endpoint, and the define the Header
API_KEY = "Your API KEY"
ENDPOINT = 'https://api.yelp.com/v3/businesses/search'
HEADERS = {'Authorization': 'bearer %s' % API_KEY}
offset = 100
for i in range(48):
    offset = 100 + 50* i
    #define the parameters
    PARAMETERS = {'location':'Atlanta',
              'offset':offset,
             'limit':50}
    # Make a request to yelp API
    response = requests.get(url = ENDPOINT, params = PARAMETERS, headers = HEADERS)
    
    # convert the JSON string to a Dictionary
    business_data = response.json()
    
    # Store the Dictionary into DataFrame
    df = pd.DataFrame.from_dict(business_data['businesses'])
    
    df_business = pd.concat([df_business,df],ignore_index = True)
    
```

```python
## Data PreProcessing
lat = []
long = []

df_cleaned = df_business[['name','rating','review_count','coordinates','categories','id','location','price']]

for i in range(np.shape(df_cleaned)[0]):
    temp_lat = df_cleaned['coordinates'][i]['latitude']
    # Concat two lists
    lat = lat + [temp_lat]
    
    temp_long = df_cleaned['coordinates'][i]['longitude']
    # Concat two lists
    long = long + [temp_long]
df_cleaned['latitude'] = lat
df_cleaned['longitude'] = long
    
df_cleaned = df_cleaned[['name','rating','review_count','categories','id','location','price','latitude','longitude']]    
```
